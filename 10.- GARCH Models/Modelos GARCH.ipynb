## **Características de las series de rendimientos financieros**


Modelar series financieras es un problema complejo. La complejidad no es solamente debido a la variedad de series (acciones, tipos de cambio, tasas de interés, etc.), a la frecuencia de medición de las observaciones (segundos, minutos, horas, etc.), o de la disponibilidad de una amplia cantidad de datos. Dicha complejidad es principalmente debido a la falta de regularidad estadística que son comunes en un gran número de series de tiempo, por lo que no resulta fácil su reproducción utilizando modelos estocásticos.

Sea $p_t$ el precio de un activo en un tiempo t y $r_t=log(p_t/p_{t-1})$ el rendimiento compuesto continuo o rendimiento logarítmico (también llamado simplemente rendimiento), la serie $r_t$ es cercana al rendimiento simple $y_t=\frac{p_t-p_{t-1}}{p_{t-1}}$, de ahí que $y_t=log(1+r_t)$. En contraste con el precio, los rendimientos no dependen de unidades monetarias lo cual facilita la comparación entre activos.

Algunas características importantes que se observan en las series de rendimientos son:


- **No estacionariedad de la serie de precios**. La serie de precios es generalmente cercana a un camino aleatorio sin intercepto, por ejemplo, la serie de rendimientos del índice de Francia CAC 40  (Cotation Assist\'ee en Continu), del $02$ de marzo de $1990$ al $15$ de octubre de $2008$, mostrada en la siguienteFigura. Por otro lado, la serie de rendimientos es generalmente compatible con la suposición de estacionariedad de segundo orden, ver Figura sobre los rendimientos del índice CAC $40$.

fig5.png
- **Ausencia de autocorrelación para la serie de rendimientos**. La serie de rendimientos generalmente muestra una baja autocorrelación, cercano a ruido blanco. La Figura muestra la función de autocorrelación simple de los rendimientos del índice CAC $40$.

- **Autocorrelación significativa del cuadrado de los rendimientos**. Los cuadrados de los rendimientos generalmente están fuertemente autocorrelacionados, ver Figura de la función de autocorrelación del cuadrado de los rendimientos del índice CAC $40$. Esta propiedad no es incompatible con la suposición de ruido blanco para los rendimientos. 

fig4.png
- **Distribución de colas pesadas**. Es común observar que la distribución empírica de los rendimientos no se ajusta a una distribución normal. Una prueba de normalidad  llevará a rechazarla. Más precisamente, las distribuciones que tienen colas pesadas son conocidas como leptocúrticas. Una medida de la leptocurtosis es el coeficiente de curtosis definido como:
$$k = \frac{{{\mu _4}}}{{{\sigma ^4}}}$$
Asintóticamente igual a $3$ para observaciones normales independientes e idénticamente distribuidas. Por independencia debe entenderse que las observaciones no esten correlacionadas e idénticamente distribuidas que sigan una distribución normal con media $\mu$ y varianza $\sigma^2$. Este coeficiente es mayor a $3$ para una serie de rendimientos.

fig8.png
- **Conglomerado de volatilidades**. Es común observar que la serie de rendimientos tiende a parecer en conglomerados. Es decir que subperiodos de turbulencia (alta volatilidad) son seguidos por subperiodos de calma (baja volatilidad). Estos periodos no son recurrentes en forma periódica (con lo cual se pueda contradecir la suposición de estacionariedad). 
- **Estacionalidad**. Los días de la semana próximos a días feriados, entre otros, pueden tener un efecto significativo en la variabilidad de los rendimientos. Seguido de un periodo de mercado cerrado, la volatilidad tiende a incrementar, reflejando la información acumulada durante el periodo de receso. Así mismo, el efecto de la estacionalidad está también presente en la serie intra-diaria.

# **Modelos ARCH**

Los modelos **ARCH** introducidos por Engle en $1982$, buscan
relajar la hipótesis de normalidad y perminten que tengamos
procesos de ruido blanco formado por variables dependientes.
Una clase de modelos más general corresponde a los modelos **GARCH**.


**Definición:**

Sea $\{y_t\}_{t \in \mathbb{N}}$ un modelo autorregresivo heterocedástico condicional denotado como $ARCH(p)$

\begin{eqnarray*}
y_t  & = & \sigma_t \varepsilon_t \\
\sigma_t^2  & = & \alpha_0 + \sum_{j=1}^p \alpha_p y_{t-j}^2
\end{eqnarray*}

$\alpha_0 > 0$, $\alpha_j \geq 0$ , con $j \in \{1,2,3,..,p\}$.  


# **Modelos GARCH**

**Definición:**

Sea $\{y_t\}_{t \in \mathbb{N}}$ un modelo autorregresivo generalizado heteroced\'astico condicional, el cual se denota como $GARCH(p,q)$

\begin{eqnarray*}
y_t  & = & \sigma_t \varepsilon_t \\
\sigma_t^2  & = & \alpha_0 + \sum_{i=1}^p \alpha_p \left(y_{t-i}^2 \right) + \sum_{j=1}^q \beta_q  \left(\sigma_{t-j}^2 \right)
\end{eqnarray*}

$\alpha_0 > 0$, $\alpha_i \geq 0$ , con $i \in \{1,2,3,..,p\}$, y  $\beta_j \geq 0$ , con $j \in \{1,2,3,..,q\}$. Adem\'as las $\alpha's$ y $\beta's$ deben cumplir ciertas condiciones para que la varianza de $\{y_t\}_{t \in \mathbb{N}}$ sea finita.  


**Modelo GARCH($1$,$1$)**

Sea $\{y_t\}_{t \in \mathbb{N}}$ un modelo  $GARCH(1,1)$

\begin{eqnarray*}
y_t  & = & \sigma_t \varepsilon_t \\
\sigma_t^2  & = & \alpha_0 +  \alpha_1 y_{t-1}^2 +  \beta_1  \sigma_{t-1}^2
\end{eqnarray*}

$\alpha_0 > 0$, $\alpha_1 \geq 0$ , y  $\beta_1 \geq 0$ , con $\{\varepsilon_t\}_{t \in \mathbb{N}}$ son v.a.i.i.d con media cero y varianza uno, entonces; $\{y_t\}_{t \in \mathbb{N}}$ es ruido blanco centrado en cero y varianza dada por $\sigma^2 = \frac{\alpha_0}{1-(\alpha_1 + \beta_j)}$ si $\alpha_j + \beta_j < 1$.

**GARCH visto como un ARMA**



\begin{eqnarray*}
y_t^2  & = & \sigma_t^2 + (y_t^2 - \sigma_t^2)\\
  & = & \alpha_0 + \alpha_1 y_{t-1}^2 + \beta_{t-1} \sigma_{t-1}^2 + y_t^2 - \sigma_t^2\\
  & = & \alpha_0 + \alpha_1 y_{t-1}^2 + \beta_{t-1} \sigma_{t-1}^2 + y_t^2 - \sigma_t^2 + \beta_{t-1}^2 y_{t-1}^2 - \beta_{t-1}^2 y_{t-1}^2\\
  & = & \alpha_0 + (\alpha_1 + \beta_1)\sigma_{t-1}^2 - (y_{t-1}^2 - \sigma_{t-1}^2) \beta_{t-1}^2 + (y_{t}^2 - \sigma_{t}^2)
\end{eqnarray*}

considerando la variable  $u_t = y_t^2 - \sigma_t^2$, se tiene que 

\begin{eqnarray*}
y_t^2 & = & \alpha_0 + (\alpha_1 + \beta_1)y_{t-1}^2 + u_t - \beta_1 u_{t-1} 
\end{eqnarray*}

podemos notar que el modelo **GARCH** admite representación **ARMA** sobre $\{y_t^2\}_{t \in \mathbb{N}}$.  

Para el caso genaral, si suponemos las mismas condiciones, $\{\varepsilon_t\}_{t \in \mathbb{N}}$ son v.a.i.i.d con media cero y varianza uno,  el modelo $GARCH(p,q)$ puede expresarse como: 

\begin{eqnarray*}
y_t^2 & = & \alpha_0 + \sum_{i=1}^{r=max(p,q)}(\alpha_i + \beta_i)y_{t-i}^2 - u_t - \sum_{j=1}^{p} \beta_{j} u_{t-j}
\end{eqnarray*}

donde $\{y_t^2\}_{t \in \mathbb{N}}$ es un $ARMA(r,p)$

## **Estimación**

Bajo la suposicición de normalidad, la función de verosimilitud para un modelo $GARCH(1,1)$ puede aproximarse mediante

\begin{eqnarray*}
f(y_1,...,y_T | \alpha, \beta) & = & f(y_1) f(y_2| F_1) f(y_3 | F_2) .... f(y_T | F_{T-1})   
\end{eqnarray*}
dado que 
\begin{eqnarray*}
f(y_T | F_{T-1}) \sim \frac{1}{\sqrt{2 \pi \sigma_t^2}} \exp\{ - \frac{y_t^2}{2\sigma_t^2} \}
\end{eqnarray*}
la funci\'on de verosimilitud condicionada a $y_1$ viene dada por
\begin{eqnarray*}
L(y_2,...,y_T | \alpha, \beta,y_1) &=& \prod_{t=2}^{T} \frac{1}{\sqrt{2\pi \sigma_t^2}} \exp\{ - \frac{y_t^2}{2\sigma_t^2} \}
\end{eqnarray*}

obteniendo el logar\'itmo
\begin{eqnarray*}
ln L(y_2,...,y_T | \alpha, \beta,y_1) & = & \sum_{2}^{T} \left[- \frac{1}{2} ln(2\pi) - \frac{1}{2} ln(\sigma_t^2) - \frac{1}{2} \frac{y_t^2}{\sigma_t^2} \right]
\end{eqnarray*}
dado que  $\frac{1}{2} ln(2\pi)$ no involucra ning\'un par\'ametro, puede omitirse
\begin{eqnarray*}
ln L(y_2,...,y_T | \alpha, \beta,y_1) & = & \sum_{2}^{T} \left[- \frac{1}{2} ln(\sigma_t^2) - \frac{1}{2} \frac{y_t^2}{\sigma_t^2} \right]
\end{eqnarray*}

Para el caso general $GARCH(p,q)$, puede verificarse La funci\'on log verosimilitud viene dada por
\begin{eqnarray*}
ln L(y_{m+1},...,y_T | \alpha, \beta, y_1,...,y_m) & = & \sum_{m+1}^{T} \left[- \frac{1}{2} ln(2\pi) - \frac{1}{2} ln(\sigma_t^2) - \frac{1}{2} \frac{y_t^2}{\sigma_t^2} \right]
\end{eqnarray*}

y al igual que en el caso anterior, puede omitirse  $\frac{1}{2} ln(2\pi)$ 

\begin{eqnarray*}
ln L(y_{m+1},...,y_T | \alpha, \beta, y_1,...,y_m) & = & \sum_{m+1}^{T} \left[- \frac{1}{2} ln(\sigma_t^2) - \frac{1}{2} \frac{y_t^2}{\sigma_t^2} \right]
\end{eqnarray*}

# **Predicción**

Dado que conocemos la información al tiempo $t$, la predicción una ventana de tiempo adelante $t+1$ viene dada por
\begin{eqnarray*}
\sigma_{t+1}^2 & = & \alpha_0  + \alpha_1 y_t^2 + \beta_t \sigma_t^2
\end{eqnarray*}

para múltiples ventanas de tiempo hacia adelante, estimamos $y_t^2$ con $E\left[y_t^2 | F_t \right]= E\left[\sigma_t^2 | F_t \right]E\left[u_t^2 | F_t \right] = \sigma_t^2$, entonces
\begin{eqnarray*}
\sigma_{t+2}^2 & = & \alpha_0  + \alpha_1 \sigma_{t+1}^2 + \beta_{1}^2 \sigma_{t+1}^2 \\
 & = & \alpha_0  + (\alpha_1 + \beta_1) \sigma_{t+1}^2 
\end{eqnarray*}
